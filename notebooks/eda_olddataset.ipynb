```python
df['Ticket Type'].value_counts()
```

```python

sns.countplot(data=df, x="Ticket Type", palette="viridis")


plt.title("Distribution of Ticket Types", fontsize=16)
plt.xlabel("Ticket Type", fontsize=12)
plt.ylabel("Count", fontsize=12)


plt.xticks(rotation=45, ha="right")

# Display the plot safely without clipping labels
plt.tight_layout()
plt.show()
```

<font color='13ac79'><h4>With this clearly we can see <b>[ticket type]</b> is evenly distributed</h4></font>
